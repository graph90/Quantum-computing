```python
# This bernstein vazirani algo allows for guessing of a binary number
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
sNum = '11'
circuit = QuantumCircuit(len(sNum)+1, len(sNum))

circuit.h(range(len(sNum)))
circuit.x(len(sNum))
circuit.h(len(sNum))
circuit.barrier()

for ii, yesno in enumerate(reversed(sNum)):
    if yesno == '1':
        circuit.cx(ii, len(sNum))


circuit.barrier()
circuit.h(range(len(sNum)))
circuit.barrier()
circuit.measure(range(len(sNum)), range(len(sNum)))
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots = 1).result()
counts = result.get_counts()
print(counts)
```
